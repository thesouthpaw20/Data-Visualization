```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data generation
np.random.seed(0)
dates = pd.date_range('2023-01-01', periods=100)
values = np.random.randn(100).cumsum()

# Create a DataFrame
df = pd.DataFrame({'Date': dates, 'Value': values})

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Value'], marker='o', linestyle='-', color='b', label='Value')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Sample Time Series Plot')
plt.legend()

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data generation
np.random.seed(0)
dates = pd.date_range('2023-01-01', periods=100)
values = np.random.randn(100).cumsum()

# Create a DataFrame
df = pd.DataFrame({'Date': dates, 'Value': values})

# 1. Line Plot
plt.figure(figsize=(12, 6))
plt.subplot(3, 3, 1)  # Subplot for line plot
plt.plot(df['Date'], df['Value'], marker='o', linestyle='-', color='b')
plt.title('Line Plot')
plt.xlabel('Date')
plt.ylabel('Value')

# 2. Area Plot
plt.subplot(3, 3, 2)  # Subplot for area plot
plt.fill_between(df['Date'], df['Value'], color='skyblue', alpha=0.4)
plt.plot(df['Date'], df['Value'], color='Slateblue', alpha=0.6)
plt.title('Area Plot')
plt.xlabel('Date')
plt.ylabel('Value')

# 3. Bar Plot
plt.subplot(3, 3, 3)  # Subplot for bar plot
plt.bar(df['Date'], df['Value'], color='green')
plt.title('Bar Plot')
plt.xlabel('Date')
plt.ylabel('Value')

# 4. Histogram
plt.subplot(3, 3, 4)  # Subplot for histogram
plt.hist(df['Value'], bins=20, color='orange', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

# 5. Box Plot
plt.subplot(3, 3, 5)  # Subplot for box plot
plt.boxplot(df['Value'])
plt.title('Box Plot')
plt.ylabel('Value')

# 6. Heatmap
plt.subplot(3, 3, 6)  # Subplot for heatmap
plt.hexbin(df.index, df['Value'], gridsize=20, cmap='YlGnBu')
plt.colorbar(label='count in bin')
plt.title('Heatmap')
plt.xlabel('Date Index')
plt.ylabel('Value')

# 7. Violin Plot
plt.subplot(3, 3, 7)  # Subplot for violin plot
plt.violinplot(df['Value'], vert=False)
plt.title('Violin Plot')
plt.xlabel('Value')
plt.yticks([])  # Disable y ticks

# 8. Scatter Plot
plt.subplot(3, 3, 8)  # Subplot for scatter plot
plt.scatter(df.index, df['Value'], color='red', marker='.')
plt.title('Scatter Plot')
plt.xlabel('Date Index')
plt.ylabel('Value')

# 9. Stacked Area Plot
plt.subplot(3, 3, 9)  # Subplot for stacked area plot
plt.stackplot(df['Date'], df['Value'], labels=['Value'], colors=['magenta'], alpha=0.5)
plt.title('Stacked Area Plot')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()

# Adjust layout and display
plt.tight_layout()
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.5, wspace=0.5)
plt.show()
```
