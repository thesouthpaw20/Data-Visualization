```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data (normal distribution)
np.random.seed(0)
data = np.random.randn(1000)

# Create histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data (normal distribution)
np.random.seed(0)
data = np.random.randn(1000)

# Create a basic histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, cumulative=True, edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Cumulative Frequency')
plt.title('Cumulative Histogram')
plt.grid(True)
plt.show()
```

```python
data1 = np.random.randn(1000)
data2 = np.random.randn(1000)

plt.figure(figsize=(8, 6))
plt.hist([data1, data2], bins=30, stacked=True, edgecolor='black', label=['Data1', 'Data2'])
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Stacked Histogram')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, density=True, edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('Normalized Histogram')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.randn(1000)

# Define custom bin edges (example: wider bins for higher values)
bins = np.arange(-3, 3, 0.5)  # Bins from -3 to 3 with width 0.5
bins = np.append(bins, np.arange(3, 6, 1))  # Wider bins for values above 3

# Create the histogram
plt.hist(data, bins=bins)
plt.xlabel("Data values")
plt.ylabel("Frequency")
plt.title("Histogram with custom bin widths")
plt.show()
```

```python
data_log = np.random.exponential(scale=1.0, size=1000)

plt.figure(figsize=(8, 6))
plt.hist(data_log, bins=30, edgecolor='black')
plt.xlabel('Values (log scale)')
plt.ylabel('Frequency')
plt.title('Histogram with Logarithmic Scale')
plt.yscale('log')
plt.grid(True)
plt.show()
```
