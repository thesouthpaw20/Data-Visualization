```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(0)
data = np.random.randn(100, 3)

# Create a basic box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['A', 'B', 'C'])
plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Basic Box Plot')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['A', 'B', 'C'], vert=False)
plt.xlabel('Values')
plt.ylabel('Groups')
plt.title('Horizontal Box Plot')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['A', 'B', 'C'], notch=True, patch_artist=True, boxprops=dict(facecolor='skyblue', color='black'))
plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Notched Box Plot')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['A', 'B', 'C'], whis=[10, 90])
plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Box Plot with Custom Whiskers')
plt.grid(True)
plt.show()
```
