```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (list of numbers)
data = [23, 45, 56, 67, 89, 34, 55, 67, 78, 89, 90, 34, 56, 78, 98, 12, 43, 65, 87, 99]

# Convert data to a pandas Series for easier manipulation
s = pd.Series(data)

# Create stem-and-leaf plot
plt.figure(figsize=(8, 6))
plt.stem(s)
plt.xlabel('Stems')
plt.ylabel('Leaves')
plt.title('Stem-and-Leaf Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate random data (100 values) with a wider range
np.random.seed(0)
data = np.random.randint(1, 1000, 100)

# Convert data to a pandas Series for easier manipulation
s = pd.Series(data)

# Create stem-and-leaf plot
plt.figure(figsize=(12, 8))
plt.stem(s)
plt.xlabel('Stems')
plt.ylabel('Leaves')
plt.title('Stem-and-Leaf Plot')
plt.grid(True)
plt.show()
```
