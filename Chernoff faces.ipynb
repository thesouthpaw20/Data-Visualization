```python
import matplotlib.pyplot as plt
import numpy as np

# Function to create Chernoff faces
def chernoff_faces(data, labels=None, scale=1.0, figsize=(8, 8)):
    # Number of variables (features) and number of faces (data points)
    n_vars = data.shape[1]
    n_faces = data.shape[0]

    # Set up the figure and axis
    fig, ax = plt.subplots(figsize=figsize)
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title('Chernoff Faces')

    # Define ranges for facial features (normalized to 0-1)
    feature_ranges = [(0.1, 0.9)] * n_vars

    # Loop through each data point (face)
    for i in range(n_faces):
        # Generate facial features
        features = data[i]
        y = 10 - (i + 1) * 10 / n_faces

        # Draw each feature of the face
        for j in range(n_vars):
            x = (j + 1) * 10 / (n_vars + 1)
            w = h = 0.8 * 10 / (n_vars + 1)

            # Normalize feature value to the range of the feature
            feature_range = feature_ranges[j]
            val = features[j]
            feature_min, feature_max = feature_range
            norm_val = (val - feature_min) / (feature_max - feature_min)

            # Draw the feature as a shape
            if labels and labels[j] == 'hair':
                plt.bar(x, y, w, norm_val * h, color='black', alpha=0.8)
            elif labels and labels[j] == 'eyes':
                plt.scatter(x + w / 2, y + h / 2, s=100, c='white', edgecolors='black', linewidths=2)
                plt.scatter(x + w * (1 - norm_val) + w / 2, y + h / 2, s=100, c='white', edgecolors='black', linewidths=2)
            elif labels and labels[j] == 'nose':
                plt.scatter(x + w / 2, y + h * (1 - norm_val) + h / 2, s=150, marker='o', c='red', edgecolors='black', linewidths=2)
            elif labels and labels[j] == 'mouth':
                plt.scatter(x + w / 2, y + h * 0.2, s=150, marker='o', c='black', edgecolors='black', linewidths=2)
            else:
                plt.scatter(x + w / 2, y + h / 2, s=150 * norm_val, marker='o', c='white', edgecolors='black', linewidths=2)

    plt.show()

# Example usage:
# Generate random data for 5 faces with 4 features each (hair, eyes, nose, mouth)
np.random.seed(0)
data = np.random.rand(5, 4)

# Define feature labels (optional)
feature_labels = ['hair', 'eyes', 'nose', 'mouth']

# Create Chernoff faces plot
chernoff_faces(data, labels=feature_labels, scale=1.5, figsize=(10, 6))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to create Chernoff faces
def chernoff_faces(data, figsize=(8, 8)):
    # Number of variables (features) and number of faces (data points)
    n_vars = data.shape[1]
    n_faces = data.shape[0]

    # Set up the figure and axis
    fig, ax = plt.subplots(figsize=figsize)
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title('Chernoff Faces')

    # Loop through each data point (face)
    for i in range(n_faces):
        # Generate facial features
        features = data[i]
        y = 10 - (i + 1) * 10 / (n_faces + 1)

        # Draw each feature of the face
        for j in range(n_vars):
            x = (j + 1) * 10 / (n_vars + 1)
            w = h = 0.8 * 10 / (n_vars + 1)

            # Normalize feature value to the range of the feature
            val = features[j]

            # Draw the feature as a shape
            if j == 0:  # Hair length
                plt.bar(x, y, w, val * h, color='black', alpha=0.8)
            elif j == 1:  # Eye size
                plt.scatter(x + w / 2, y + h / 2, s=100 * val, c='white', edgecolors='black', linewidths=2)
            elif j == 2:  # Nose size
                plt.scatter(x + w / 2, y + h * 0.6, s=150 * val, marker='o', c='red', edgecolors='black', linewidths=2)
            elif j == 3:  # Mouth size
                plt.scatter(x + w / 2, y + h * 0.2, s=150 * val, marker='o', c='black', edgecolors='black', linewidths=2)

    plt.show()

# Example usage:
# Generate random data for 5 faces with 4 features each (hair, eyes, nose, mouth)
np.random.seed(0)
data = np.random.rand(5, 4)

# Create Chernoff faces plot
chernoff_faces(data, figsize=(10, 6))
```
