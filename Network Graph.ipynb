```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a sample graph for Bible references
G = nx.Graph()

# List of all 66 books of the Bible
books = [
    'Genesis', 'Exodus', 'Leviticus', 'Numbers', 'Deuteronomy', 'Joshua', 'Judges', 'Ruth', 
    '1 Samuel', '2 Samuel', '1 Kings', '2 Kings', '1 Chronicles', '2 Chronicles', 'Ezra', 
    'Nehemiah', 'Esther', 'Job', 'Psalms', 'Proverbs', 'Ecclesiastes', 'Song of Solomon', 
    'Isaiah', 'Jeremiah', 'Lamentations', 'Ezekiel', 'Daniel', 'Hosea', 'Joel', 'Amos', 
    'Obadiah', 'Jonah', 'Micah', 'Nahum', 'Habakkuk', 'Zephaniah', 'Haggai', 'Zechariah', 
    'Malachi', 'Matthew', 'Mark', 'Luke', 'John', 'Acts', 'Romans', '1 Corinthians', 
    '2 Corinthians', 'Galatians', 'Ephesians', 'Philippians', 'Colossians', '1 Thessalonians', 
    '2 Thessalonians', '1 Timothy', '2 Timothy', 'Titus', 'Philemon', 'Hebrews', 'James', 
    '1 Peter', '2 Peter', '1 John', '2 John', '3 John', 'Jude', 'Revelation'
]

G.add_nodes_from(books)

# Add major cross-references between books
edges = [
    ('Genesis', 'Exodus'), ('Genesis', 'Job'), ('Exodus', 'Numbers'), ('Exodus', 'Deuteronomy'), 
    ('Numbers', 'Deuteronomy'), ('Deuteronomy', 'Joshua'), ('Joshua', 'Judges'), ('Judges', 'Ruth'), 
    ('Ruth', '1 Samuel'), ('1 Samuel', '2 Samuel'), ('2 Samuel', '1 Kings'), ('1 Kings', '2 Kings'), 
    ('2 Kings', 'Isaiah'), ('Isaiah', 'Jeremiah'), ('Jeremiah', 'Lamentations'), ('Lamentations', 'Ezekiel'), 
    ('Ezekiel', 'Daniel'), ('Daniel', 'Hosea'), ('Hosea', 'Joel'), ('Joel', 'Amos'), ('Amos', 'Obadiah'), 
    ('Obadiah', 'Jonah'), ('Jonah', 'Micah'), ('Micah', 'Nahum'), ('Nahum', 'Habakkuk'), ('Habakkuk', 'Zephaniah'), 
    ('Zephaniah', 'Haggai'), ('Haggai', 'Zechariah'), ('Zechariah', 'Malachi'), ('Matthew', 'Mark'), 
    ('Mark', 'Luke'), ('Luke', 'John'), ('John', 'Acts'), ('Romans', '1 Corinthians'), ('1 Corinthians', '2 Corinthians'), 
    ('2 Corinthians', 'Galatians'), ('Galatians', 'Ephesians'), ('Ephesians', 'Philippians'), ('Philippians', 'Colossians'), 
    ('Colossians', '1 Thessalonians'), ('1 Thessalonians', '2 Thessalonians'), ('2 Thessalonians', '1 Timothy'), 
    ('1 Timothy', '2 Timothy'), ('2 Timothy', 'Titus'), ('Titus', 'Philemon'), ('Hebrews', 'James'), ('James', '1 Peter'), 
    ('1 Peter', '2 Peter'), ('2 Peter', '1 John'), ('1 John', '2 John'), ('2 John', '3 John'), ('3 John', 'Jude'), 
    ('Jude', 'Revelation')
]

G.add_edges_from(edges)

# Draw the network graph
plt.figure(figsize=(18, 12))  # Adjust figure size as needed
pos = nx.spring_layout(G, seed=42)  # positions for all nodes, seed for reproducibility

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=700, edgecolors='k', linewidths=1.5)

# Draw edges
nx.draw_networkx_edges(G, pos, edgelist=edges, width=1.0, alpha=0.5, edge_color='gray')

# Draw labels
nx.draw_networkx_labels(G, pos, font_size=8, font_color='black', verticalalignment='top', horizontalalignment='center')

# Add a title
plt.title('Major Bible References Network Graph', fontsize=15, fontweight='bold')

# Remove axis
plt.axis('off')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Sample nodes and edges
nodes = ['A', 'B', 'C', 'D']
edges = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'A')]

# Create a graph object
G = nx.Graph()

# Add nodes to the graph
G.add_nodes_from(nodes)

# Add edges to the graph
G.add_edges_from(edges)

# Generate a layout for the nodes
pos = nx.spring_layout(G)

# Draw the nodes
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=700, edgecolors='k', linewidths=1.5)

# Draw the edges
nx.draw_networkx_edges(G, pos, edgelist=edges, width=2.0, alpha=0.5, edge_color='gray')

# Draw the labels
nx.draw_networkx_labels(G, pos, font_size=12, font_color='black', font_family='sans-serif')

# Add a title
plt.title('Sample Network Graph', fontsize=15, fontweight='bold')

# Display the plot
plt.axis('off')  # Turn off the axis
plt.show()
```
